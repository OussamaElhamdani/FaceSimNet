## Historical Context
In this stage of our notebook, you get to familiarize yourself with some of the history of the big aspects in computer vision to understand the approach which simplify the problem
In the history of deep learning, the two most powerful tools that will help us in this notebook are:

* Transfer Learning
*  Embeddings Learning

It's worth noting that transfer learning has been always been used in computer vision, since it does reduce the computations and the gradients that must be calculated during backpropagation. On the other hand, embeddings learning is what really defines the idea of similarity: At the early days of deep learning marked a shift in pattern recognition, in fact deep neural networks shows a remarkable capabilities in feature learning. The problem with these early models was the size of the weights matrices since the modeling of images in these days was with a flatten vectors, later the CNNs then emerged to address this problem in Models like AlexNet, VGGNet, and ResNet achieved unprecedented accuracy in image classification tasks.

In order to understand the idea of similarity and embeddings, let's dive into the simplest to understand and closest problem of our FaceSimNet, Face recognition which is a famous problem in computer vision that aims to recognize wheter or not an input is present in a database for example. To build such a model with only what we've covered before a major problem can occur:
* One-Shot learning

This problem is data-related, the problem states to learn the task despite the limitation of size of the data set because in most face recognition systems you might have one to few images of the same person, and the deep learning models we've been talking about does not do really well with one or few training examples. And even the naive solution that comes up to head at first glace when hearing this, that is to say Input layer $->$ Some Conv and Pooling layers $->$ softmax output unit, this can't work simply because of the possibility of adding new different images to the database.

To address this problem of recognizing the person again from only one training example $($Worst case$)$ the idea of constructing a network that learns the similarity of two images comes up. The core of the idea is to build a neural network that can give an encoding of an images (Embeddings), and so if two images are the same these encodings should be really close, so formally creating a net that learns a vector space to represent images.

One way to learn the parameters of a neural network so that it can learn the similarity between two images is:

* Triplet Loss: $L(A, P, N) = max (0, d(A, P) − d(A, N) + α)$
The triplet loss ensures that the distance between an anchor (A) and a
positive example (P) in the embedding space is minimized, while the
distance between the anchor and a negative example (N) is maximized.
